```python
import pandas as pd
import numpy as np
import regex as re
import json
```

```python
data_v4 = pd.read_pickle('./preprocessed_alcohol_v4.pkl')
data_v3 = pd.read_pickle('./preprocessed_alcohol_v3.pkl')
```

```python
columns = data_v4.columns
print(columns)
train_data = data_v4.drop(columns=['ModifiedDate','ImageUrlPath', 'IBA'], axis=1) # 필요하지 않는 열은 드랍합니다.
```

```python
train_data['DrinkName']

def to_lower(row):
    return row.lower()
```

```python
# match Ingredient index and Measurement index
for i in range(len(train_data)):
    ingredient = train_data['Ingredients'][i]
    measurement = train_data['Measures'][i]
    ingredient.sort(key= lambda x: x[1])
    measurement.sort(key= lambda x: x[1])
```

```python
for i in range(len(train_data)):
    for j in range(len(train_data['Ingredients'][i])):
        ing_index = train_data['Ingredients'][i][j][1]
        mes_index = train_data['Measures'][i][j][1]
        if ing_index != mes_index:
            print('Ingredient index and Measurement index is not matched!')
            print('Index: ', i)
print('All matched!')
```

```python
train_data
```

```python
# ingredient에서 매칭 index 제거하기
for i in range(len(train_data)):
    ingredient_list = []
    for j in range(len(train_data['Ingredients'][i])):
        ingredient_list.append(train_data['Ingredients'][i][j][0])
    train_data['Ingredients'][i] = ingredient_list
    
for i in range(len(train_data)):
    measurement_list = []
    for j in range(len(train_data['Measures'][i])):
        measurement_list.append(train_data['Measures'][i][j][0])
    train_data['Measures'][i] = measurement_list
```

```python
for i in range(len(train_data)):
    print(train_data['Ingredients'][i], train_data['Measures'][i])
```

```python
train_data['Recipe'] = None
for i in range(len(train_data)):
    ingredient_measurement_dict = {}
    for j in range(len(train_data['Ingredients'][i])):
        ingredient = train_data['Ingredients'][i][j]
        ingredient_measurement_dict[ingredient] = train_data['Measures'][i][j]
    train_data['Recipe'][i] = ingredient_measurement_dict
```

```python
train_data['Recipe']
```

```python
json_data = {"cocktail_info":[]}

for i in range(len(train_data)):
    element = {}
    element['cocktail_name'] = train_data['DrinkName'][i]
    element['cocktail_glass'] = train_data['Glass'][i]
    element['category'] = train_data['Category'][i]
    element['recipe'] = train_data['Recipe'][i]
    element['alcoholic'] = train_data['Alcoholic'][i]
    json_data['cocktail_info'].append(element)
```

```python
with open('./train_data.json', 'w', encoding='utf-8') as make_file:
    json.dump(json_data, make_file)

with open('./train_data.json', 'r') as f:
    json_data = json.load(f)
```
