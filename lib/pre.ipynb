```python
import os

# Directory containing the files
# C:\Users\woo\gist\24_1\de\FeelFlask_fork\backend\data_works\drinks_img
directory = r'C:\Users\woo\gist\24_1\de\FeelFlask_fork\backend\data_works\drinks_img'

# Loop through all files in the directory
for filename in os.listdir(directory):
    # Construct full file path
    file_path = os.path.join(directory, filename)
    
    # Check if it's a file (not a directory) and if it has no file extension
    if os.path.isfile(file_path) and '.' not in filename:
        # New file path with .jpg extension
        new_file_path = f"{file_path}.jpg"
        
        # Rename the file
        os.rename(file_path, new_file_path)
        print(f"Renamed '{file_path}' to '{new_file_path}'")

print("All files without an extension have been updated to .jpg")
```
