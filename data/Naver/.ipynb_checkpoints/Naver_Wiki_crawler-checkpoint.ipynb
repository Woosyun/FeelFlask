````python

import io
import os
from tqdm import tqdm
from google.cloud import vision
````

````python
liq_list = [] 

for i in tqdm(output): #File iter 
    path = f'{i}'
    
    client = vision.ImageAnnotatorClient()
    
    # Loads the image into memory
    with io.open(path, 'rb') as image_file:
        content = image_file.read()
        image = vision.Image(content=content)
        # Performs text detection on the image file
        response = client.text_detection(image=image)
        texts = response.text_annotations
    liq = []    
    for text in texts:
        content = text.description
        content = content.split('\n')[0]
        liq.append(content)
    liq_list.append(liq)
print(liq_list)
````

````python
# 이미지 전처리 -> 이미지 흑백 전환 -> 사각형 경계 명확히 
# 사각형 위치, 크기 식별
# 객체 인식 
# 사각형 정량화 
# 데이터 추출
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Load the image from file system
image_path = 'img/img_희양산막걸리15그래프이미지1datafonti.png'
image = cv2.imread(image_path)

# Convert to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Binarize the image - we are looking for gray squares which should be lighter than the background but darker than white
_, binary_image = cv2.threshold(gray_image, 128, 255, cv2.THRESH_BINARY_INV)

# Find contours
contours, _ = cv2.findContours(binary_image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Filter out small contours that are not our squares
squares = [cnt for cnt in contours if cv2.contourArea(cnt) > 200]

# Create an empty list to store the center points of the squares
square_centers = []

# Find the center point of each square and store it in the list
for sq in squares:
    M = cv2.moments(sq)
    if M['m00'] != 0:
        cX = int(M['m10'] / M['m00'])
        cY = int(M['m01'] / M['m00'])
        square_centers.append((cX, cY))

# Show the binarized image and the center points of the squares
plt.imshow(binary_image, cmap='gray')
for center in square_centers:
    plt.scatter(center[0], center[1], c='red', s=10)
plt.show()

# Return the number of squares which will help us understand the counts for each parameter
len(squares)

   
````

````python
import numpy as np

# Convert image to grayscale
image_path = 'img/img_희양산막걸리15그래프이미지1datafonti.png'
img = Image.open(image_path)
gray_img = img.convert('L')

# Threshold to determine whether a box is filled
# We pick a value based on typical gray values, to be adjusted after inspection
threshold = 100

# The coordinates for the boxes will be determined based on visual inspection of the image
# We will assume that the boxes are regularly spaced and have the same size
# Define the positions of the rows for each parameter
# These positions are estimated manually based on the y-axis displayed in the image above
# Similarly, the x positions are estimated from the x-axis displayed in the image
row_positions = {
    'sweetness': 110,  # y position for the 'sweetness' row
    'sourness': 170,   # y position for the 'sourness' row
    'refreshing': 230, # y position for the 'refreshing' row
    'body': 290        # y position for the 'body' row
}
# We estimate there are 7 boxes, and they seem to be 50 pixels wide and spaced by around 50 pixels
box_width = 50
box_spacing = 50
# Estimate the starting x position of the first box
start_x = 75

# Now we will inspect each row and count the number of filled boxes based on the threshold
values = {}

for parameter, y in row_positions.items():
    filled_boxes = 0
    # Inspect each box in the row
    for i in range(7):
        # Calculate the x position of the current box
        x = start_x + i * (box_width + box_spacing)
        # Extract the box region from the image
        box = gray_img.crop((x, y, x + box_width, y + box_width))
        # Calculate the average intensity of the box
        avg_intensity = np.array(box).mean()
        # If the intensity is below the threshold, consider the box filled
        if avg_intensity < threshold:
            filled_boxes += 1
    # Store the number of filled boxes for the parameter
    values[parameter] = filled_boxes

values
````

````python
import cv2
import numpy as np

# Load the image
image_path = 'img/img_희양산막걸리15그래프이미지1datafonti.png'
image = cv2.imread(image_path)

# Check if image is loaded
if image is None:
    raise Exception("Image not found or path is incorrect")

# Convert the image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# We will apply a range of thresholds to find gray, assuming gray is neither too close to black nor to white
# We'll need to adjust these values based on the actual image
lower_gray = np.array([100])  # Lower bound of gray color
upper_gray = np.array([200])  # Upper bound of gray color

# Create a mask to only select the parts of the image within the gray range
gray_mask = cv2.inRange(gray_image, lower_gray, upper_gray)

# Find contours on the mask
contours, _ = cv2.findContours(gray_mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Draw bounding boxes around the detected gray boxes in the original image
for cnt in contours:
    x, y, w, h = cv2.boundingRect(cnt)
    cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 2)

# Save the result
output_path = 'detected_gray_boxes.png'
cv2.imwrite(output_path, image)

# Output the path to the processed image and the count of detected gray boxes
# output_path, 
len(contours)
````

````python

# 필요패키지 import
import time
import json
import urllib
import numpy as np
import pandas as pd
import csv
from collections import Counter
name_list = []
f = open('name_feature.csv', 'r', encoding= 'utf-8')
reader = csv.reader(f)
for row in reader:
    name_list.append(row)
f.close()  
# header가 함께 불러와졌기 때문에 삭제
del name_list[0]
# 중복값 제거 : dict.fromkeys를 이용해 순서를 지키며 중복값을 없애줌!
name_list = list(map(lambda x: list(dict.fromkeys(x)), name_list))
# 리스트 내 value count를 위해 1차원 리스트로 변환
name_list1 = sum(name_list, [])
# list 내 value count -> counter 사용
result = Counter(name_list1)
````
