##### 여러개의 measure값들 하나로 열로 통합하기

measure 값들은 각각 번호에 따라 ingredient 재료 사용량을 의미한다.


```python
# 그전에 measure값이 정확히 무엇을 의미하는지 조사할 필요가 있다.
# 인덱스 4번의 술 belmont155의 정보가 담긴 사이트를 들어가 보자. 
# https://www.thecocktaildb.com/drink/15346-155-Belmont

measure_list = []
for i in range(1, 16):
    col = 'strMeasure' + str(i)
    features = alcohol.at[4, col]
    measure_list.append(features)
print(alcohol.at[4, 'strDrink'])
print(measure_list)
print(alcohol.at[4, 'Ingredients'])
```

```python
alcohol['Measures'] = None
```

```python
for i in range(len(alcohol)):
    measure_list = []
    for j in range(1, 16):
        col = 'strMeasure' + str(j)
        features = alcohol.at[i, col]
        if not(pd.isnull(features) or features == ' ' or features == '\n' or features == ' \n' or features==None):
            measure_list.append((features, j))
    alcohol.at[i, 'Measures'] = measure_list
```

```python
drop_list = []
for j in range(1, 16):
    col = 'strMeasure' + str(j)
    drop_list.append(col)

alcohol.drop(drop_list, axis=1, inplace=True)
```

```python
# for example, measure 측정 값에서는 2 1/2와 같이 의미가 조금 불분명하거나, fill with 처럼 정량화된 값이 아닌 형태로 표현된 것도 있다.
print(alcohol['Measures'][10])
print(alcohol['strDrink'][10])
```

```python
# 1. 표시한 재료의 사용량과 대응되는 사용한 재료의 갯수가 서로 일치하지 않는다
# 2. 재료의 사용량이 표시되지 않는 일부 빈값도 존재한다. (아래 셀 확인)
# 3. 재료의 사용량 표기는 rough한 편으로 정제가 필요하다.
for i in range(len(alcohol)):
    len_ing = len(alcohol['Ingredients'][i])
    len_measure = len(alcohol['Measures'][i])
    # print(f'len_ing: {len_ing}, len_measures: {len_measure}')
    if (len_ing != len_measure):
        print(f'len_ing: {len_ing}, len_measures: {len_measure}')
```

```python
# measure된 값에서는 7개의 값이 비어있다.
index_list = []
for i in range(len(alcohol)):
    if len(alcohol['Measures'][i]) == 0:
        print('Empty Measures!')
        index_list.append(i)
# total 7 empty Measures..
test = alcohol.drop(index_list, axis=0)
test.reset_index(inplace=True, drop=True)

for i in range(len(test)):
    if len(test['Measures'][i]) == 0:
        print('test Empty Measures!')
```

```python
# 일부는 단위가 누락되어져 있고, 1 1/2와 같이 일부는 이해하기 어려운 형태로 표기되어 있다.
# 사용한 량이 누락되어 있는 경우도 있다. 또한 string 단위로 표현되 정량적인 수치 표현이 안되어 있다.
# 일부 단위는 단위 자체가 정량적인 값으로 표현되어 있지 않고, 대략적인 량만 알 수 있도록 되어 있다. (fill with나 Juice of 등)
# 일부는 add와 같은 동사도 포함되어져 있다. 또한 사용된 재료도 일부 포함되어 있는 걸로 보인다.
uni_measure = np.unique(alcohol['Measures'])
uni_measure
```

#### 전처리된 값 저장

1. NULL이나 빈값이 포함된 행은 삭제하지 않고 저장했습니다. (DateModified, strIBA, strInstruction, StrMeasure)
2. 데이터 열의 이름을 수정하여서 저장했습니다.
3. Ingredients, Measures는 (재료, index(1~15)), (사용량, index(1~15))이렇게 저장하였습니다.
4. csv 파일 대신 pickle 파일로 저장하여야 list 값이 txt로 변환되지 않고 그대로 저장됩니다.


```python
column_names = alcohol.columns
rename_names = ['DrinkName', 'ModifiedDate', 'DrinkID', 'Alcoholic', 'Category', 'ImageUrlPath', 'Glass', 'IBA', 'Instruction', 'Ingredients', 'Measures']
mapper = {}

for i, column in enumerate(column_names):
    mapper[column] = rename_names[i]

mapper
```

```python
save_alcohol = alcohol.rename(columns = mapper)
save_alcohol.to_pickle('./preprocessed_alcohol.pkl')
# save_alcohol.to_csv('./preprocessed_alcohol.csv', index=False)
```
